```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/JournalData/SampleData.csv")
```

```python
data.head()
```

```python
sentences = data.Abstract.values
```

```python
from InstructorEmbedding import INSTRUCTOR
model = INSTRUCTOR('hkunlp/instructor-xl')
```

```python
instruction = "Represent the abstract of the below research paper:"
```

```python
text = []
for i,sentence in enumerate(sentences):
  text.append([instruction, sentence])
text[0]
```

```python
import sklearn.cluster
embeddings = model.encode(text)
clustering_model = sklearn.cluster.MiniBatchKMeans(n_clusters=3)
clustering_model.fit(embeddings)
cluster_assignment = clustering_model.labels_
print(cluster_assignment)
```
